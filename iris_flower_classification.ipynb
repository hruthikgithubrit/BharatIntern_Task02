```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

Predicting the type of Iris flower after training using a ML algorithm

INPUT?
sepal length
Petal length
sepal width
Petal width

OUTPUT?
Class of flower

Loading the Dataset


```python
columns = ['sepal length','sepal width','Petal length','Petal width','Class_labels']
#Load the file
df = pd.read_csv('iris.data.csv',names=columns)
df.head(150)
```

VISUALIZATION OF OUR DATASET

```python
df.describe()
```

```python
#visualize the whole dataset
sns.pairplot(df, hue='Class_labels')
```

SEPARATING THE INPUT COLUMNS AND OUTPUT COLUMN

```python
data = df.values

x = data[:,0:4] #last column class label is removed
y = data[:,4] #only class label is present.
```

SPLITTING THE DATA INTO TRAINING AND TESTING

```python
#split the data into train and test dataset
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
print(y_train)
```

MODEL 1: Support vector machine algorithm

```python
from sklearn.svm import SVC
model_svc = SVC()
model_svc.fit(x_train, y_train)
```

```python
prediction1 = model_svc.predict(x_test)
#calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction1))
for i in range(len(prediction1)):
  print(y_test[i],prediction1[i])
```

MODEL2: LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
model_LR = LogisticRegression()
model_LR.fit(x_train, y_train)
```

```python
prediction2 = model_LR.predict(x_test)
#calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction2))
for i in range(len(prediction2)):
  print(y_test[i],prediction2[i])
```
